```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
titanic.to_csv('./titanic.csv', index = False)
#이후 저장한 titanic.csv 파일을 열어서 전처리 작업이 필요한지 확인
```

```python
titanic.isnull().sum()      #  >>>age, embarked, deck, embark_town에서 누락값 확인
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].median())     # age열의 누락값을 중앙값으로 처리
```

```python
titanic['embarked'].value_counts()      #embarked 열의 최빈값을 확인
```

```python
titanic['embarked'] = titanic['embarked'].fillna('S')       #embarked 열의 누락값을 최빈값으로 처리
```

```python
titanic['embark_town'].value_counts()     #embarked와 동일
```

```python
titanic['embark_town'] = titanic['embark_town'].fillna('Southampton')
```

```python
titanic['deck'].value_counts()
```

```python
titanic['deck'] = titanic['deck'].fillna('C')
```

```python
titanic.isnull().sum()
```

```python
titanic.info()
```

```python
titanic.survived.value_counts()     #1은 생존
```

```python
#차트로 시각화
f, ax = plt.subplots(1,2,figsize=(12,6))
titanic['survived'][titanic['sex'] == 'male'].value_counts().plot.pie(explode = [0,0.1], autopct = '%1.1f%%', ax=ax[0], shadow=True)
titanic['survived'][titanic['sex'] == 'female'].value_counts().plot.pie(explode = [0,0.1], autopct = '%1.1f%%', ax=ax[1], shadow=True)
ax[0].set_title('Survived (Male)')
ax[1].set_title('Survived (Female)')
plt.show()
```

```python
sns.countplot('pclass', hue = 'survived', data = titanic)       #객실 등급별 생존율 확인
plt.title('Pclass vs Survived')
plt.show()
```

```python
#상관분석 실시
titanic_corr = titanic.corr(method = 'pearson')
titanic_corr
```

```python
titanic_corr.to_csv('./titanic_corr.csv', index=False)
```

```python
#객실별 남녀 성별별로 생존율 도식화
sns.catplot(x = 'pclass', y = 'survived', hue = 'sex', data = titanic, kind = 'point') 
plt.show()
```

```python
#변수사이의 상관계수를 히트맵으로 시각화하기
def category_age(x):
    if x<10:
        return 0
    elif x<20:
        return 1
    elif x<30:
        return 2
    elif x<40:
        return 3
    elif x<50:
        return 4
    elif x<60:
        return 5
    elif x<70:
        return 6
    else:
        return 7
```

```python
titanic['age2'] = titanic['age'].apply(category_age)
```

```python
titanic['sex'] = titanic['sex'].map({'male':1, 'female':0})
```

```python
titanic['family'] = titanic['sibsp'] + titanic['parch'] +1
```

```python
titanic.to_csv('./titanic3.csv', index=False)
```

```python
heatmap_data = titanic[['survived','sex','age2','family','pclass','fare']]
colormap = plt.cm.RdBu
sns.heatmap(heatmap_data.astype(float).corr(), linewidths = 0.1, vmax=1.0, square=True, cmap=colormap, linecolor='white',annot=True, annot_kws={'size':10})
plt.show()
```
